## LPノルムの等高線を表示
・LP0.1からLP1000までの値の等高線を表示  

```python
# LPノルムの計算
def lp_norm(x, y, lp=1):
    x = np.abs(x) 
    y = np.abs(y)
    return (x**lp + y**lp)**(1./lp)

# LPノルムの等高線を描画
def draw_lp_contour(lp=1, xlim=(0, 1), ylim=(0, 1)):
    n = 201
    X, Y = np.meshgrid(np.linspace(xlim[0], xlim[1], n), np.linspace(ylim[0], ylim[1], n))
    Z = lp_norm(X, Y, lp) 
    interval = [i/10. -1 for i in range(20)]
    plt.contour(X, Y, Z, interval)
    plt.title("Contour of LP{}".format(lp))
    plt.xlim(xlim[0], xlim[1])
    plt.ylim(ylim[0], ylim[1])

# 16種のpの値のグラフを描画
fig =plt.figure(figsize=(14,14))
size = 4
for i, lp in enumerate(np.r_[[0.1], np.linspace(0.5, 7, 14), [1000]]):
    plt.subplot(size, size, i+1)
    draw_lp_contour(lp, (-1, 1),(-1, 1))
```

(1) (1,0)

## Lasso回帰クラス

```python
def soft_thresholding(x, y):
    return np.sign(x) * max(abs(x) - y, 0)

# Lasso回帰クラス
class Lasso:
    # コンストラクタ
    def __init__(self, lambda_, eps=0.0001, max_iter=1000):
        self.lambda_ = lambda_
        self.eps = eps
        self.max_iter = max_iter
        self.w_ = None
    # 予想
    def fit(self, X, t):
        n, d = X.shape
        self.w_ = np.zeros(d + 1)
        avgl1 = 0.
        for _ in range(self.max_iter):
            avgl1_prev = avgl1
            self._update(n, d, X, t)
            avgl1 = np.abs(self.w_).sum() / self.w_.shape[0]
            # 誤差が予測半にないに収まるまで繰り返す
            if abs(avgl1 - avgl1_prev) <= self.eps:
                break
    # 更新
    def _update(self, n, d, X, t):
        self.w_[0] = (t - np.dot(X, self.w_[1:])).sum() / n
        w0vec = np.ones(n) * self.w_[0]
        for k in range(d):
            ww = self.w_[1:]
            ww[k] = 0
            q = np.dot(t - w0vec - np.dot(X, ww), X[:, k])
            r = np.dot(X[:, k], X[:, k])
            self.w_[k + 1] = soft_thresholding(q / r, self.lambda_)
    # 予想
    def predict(self, X):
        if X.ndim == 1:
            X = X.reshape(X.shape[0], 1)
        Xtil = np.c_[np.ones(X.shape[0]), X]
        return np.dot(Xtil, self.w_)
```

## Lassoを利用した学習のサンプル
・**winequality-red.csv**を読み込んで、学習・テストを行う  
・正則加項にかける稀有数**_lambda**が大きくなるほど、パラメータ中の0の数が多くなることを確認  
・0の数が増えてもMSEに大きな変化がない

```python
import csv

# データ読み込み
Xy = []
# データはwinequality-red.csv
with open("./data/winequality-red.csv") as fp:
    for row in csv.reader(fp, delimiter=";"):
        Xy.append(row)
Xy = np.array(Xy[1:], dtype=np.float64)

# 訓練用データとテスト用データに分割する
np.random.seed(0)
np.random.shuffle(Xy)
train_X = Xy[:-1000, :-1]
train_y = Xy[:-1000, -1]
test_X = Xy[-1000:, :-1]
test_y = Xy[-1000:, -1]

# ハイパーパラメータを変えながら学習させて結果表示
for lambda_ in [0.001, 0.01 , 0.1, 1.0, 10.]:
    model = Lasso(lambda_)
    model.fit(train_X, train_y)
    y = model.predict(test_X)
    print("--- lambda = {} ---".format(lambda_))
    print("coefficients:")
    print(model.w_)
    mse = ((y - test_y)**2).mean()
    print("MSE: {:.3f}".format(mse))
```

## 線形回帰クラス

```python
from scipy import linalg

# 線形回帰
class LinearRegression:
    # コンストラクタ
    def __init__(self):
        self.w_ = None
    # 学習
    def fit(self, X, t):
        Xtil = np.c_[np.ones(X.shape[0]), X]
        A = np.dot(Xtil.T, Xtil)
        b = np.dot(Xtil.T, t)
        #self.w_ = linalg.solve(A, b)
        A_inv = np.linalg.inv(A) # Aの逆行列
        # 重みの取得
        self.w_ = A_inv.dot(b)
    # 予想
    def predict(self, X):
        if X.ndim == 1:
            X = X.reshape(1, -1)
        Xtil = np.c_[np.ones(X.shape[0]), X]
        return np.dot(Xtil, self.w_)
```

## リッジ回帰

```python
# リッジ回帰
class RidgeRegression:
    def __init__(self, lambda_=1.):
        self.lambda_ = lambda_
        self.w_ = None

    def fit(self, X, t):
        Xtil = np.c_[np.ones(X.shape[0]), X]
        c = np.eye(Xtil.shape[1])
        A = np.dot(Xtil.T, Xtil) + self.lambda_ * c
        b = np.dot(Xtil.T, t)
        A_inv = np.linalg.inv(A) # Aの逆行列
        # 重みの取得
        self.w_ = A_inv.dot(b)

    def predict(self, X):
        Xtil = np.c_[np.ones(X.shape[0]), X]
        return np.dot(Xtil, self.w_)
```

## 線形回帰とリッジ回帰の比較

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

# 乱数の種は0
np.random.seed(0)

# 乱数で点を置く
n = 31
x = np.linspace(0, n - 1, n)
y = 0.3*x  - 3 +(np.random.rand(n) - 0.5) * 10.0

# リッジ回帰による計算
model = RidgeRegression(1.)
model.fit(x, y)
b1, a1 = model.w_

# 線形回帰による計算
model = LinearRegression()
model.fit(x, y)
b2, a2 = model.w_

plt.scatter(x, y, color="k")
xmax = x.max()
# リッジ回帰の直線を表示
plt.plot([0, xmax], [b1, b1 + a1 * xmax], color="r" ,label="Ridge")
plt.plot([0, xmax], [b2, b2 + a2 * xmax], color="g",label="Linear")
plt.legend()
plt.show()
```
